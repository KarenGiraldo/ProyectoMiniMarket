```python
# Definir las columnas que deseas modificar
columnas_a_modificar = [
    'especificidad_oncologico', 'especificidad_trasplante_osteomuscular',
    'especificidad_trasplante_piel', 'especificidad_trasplante_cardiovascular',
    'especificidad_trasplante_tejido_ocular', 'especificidad_atencion_paciente_quemado',
    'especificidad_salud_mental', 'especificidad_spa', 'especificidad_otras_patologias',
    'especificidad_trasplante_celulas_progenitoras_hematopoyeticas',
    'especificidad_procedimientos_quirurgicos_ambulatorios', 'especificidad_organo_rinon',
    'especificidad_organo_higado', 'especificidad_organo_pancreas', 'especificidad_organo_intestino',
    'especificidad_organo_multivisceral', 'especificidad_organo_corazon',
    'especificidad_organo_pulmon', 'especificidad_sustancias_psicoactivas',
    'especificidad_trasplante_renal'
]

# Reemplazar "Sí" por 1 y "No" por 0 en las columnas especificadas
servicios[columnas_a_modificar] = servicios[columnas_a_modificar].replace({'SI': 1, 'NO': 0})

# Reemplazar los valores nulos por 999
servicios[columnas_a_modificar] = servicios[columnas_a_modificar].fillna(999)

# Convertir todas las columnas especificadas a enteros
servicios[columnas_a_modificar] = servicios[columnas_a_modificar].astype(int)
```

```python
# Agrupar por 'codigo_sede' y sumar los valores de las columnas especificadas
servicios2 = servicios.groupby('codigo_sede').agg({
    'oferta_oncologia': 'sum',
    'oferta_transplantes': 'sum',
    'oferta_quemados': 'sum',
    'oferta_salud_mental': 'sum',
    'oferta_cirugia_ambulatoria': 'sum',
    'oferta_patologia_especifica': 'sum',
    'oferta_enfasis': 'sum'
}).reset_index()

# Convertir la suma a 1 si es mayor a 0, de lo contrario dejar en 0
columns_to_update = ['oferta_oncologia', 'oferta_transplantes', 'oferta_quemados', 
                      'oferta_salud_mental', 'oferta_cirugia_ambulatoria', 
                      'oferta_patologia_especifica', 'oferta_enfasis']

servicios2[columns_to_update] = servicios2[columns_to_update].applymap(lambda x: 1 if x > 0 else 0)

# Convertir la columna 'codigo_sede' a índice
servicios2 = servicios2.set_index('codigo_sede')

servicios2
```
